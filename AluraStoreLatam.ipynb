### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
dataframes = [tienda, tienda2, tienda3, tienda4]
todo_junto = pd.concat(dataframes, ignore_index=True)

total_facturacion = todo_junto['Precio'].sum()
print(f"La facturación total de todas las tiendas es: {total_facturacion:,.2f}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria = todo_junto.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_por_categoria = ventas_por_categoria.sort_values(by='Precio', ascending=False)
print("Ventas por Categoría de Producto:")
print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = todo_junto['Calificación'].mean()
print(f"La calificación promedio de la tienda es: {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = todo_junto['Producto'].value_counts()

print("Productos más vendidos (Top 5):")
print(productos_vendidos.head(5))

print("\nProductos menos vendidos (Bottom 5):")
print(productos_vendidos.tail(5))
```

# 5. Envío promedio por tienda

```python
print("Envío promedio por tienda:")
print(f"Tienda 1: {tienda['Costo de envío'].mean():,.2f}")
print(f"Tienda 2: {tienda2['Costo de envío'].mean():,.2f}")
print(f"Tienda 3: {tienda3['Costo de envío'].mean():,.2f}")
print(f"Tienda 4: {tienda4['Costo de envío'].mean():,.2f}")
```
